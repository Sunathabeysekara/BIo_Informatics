```python
control=dataset['Gene']
control=pd.concat([control,dataset.iloc[:,1:11]],axis=1)
```

```python
treatment=dataset['Gene']
treatment=pd.concat([treatment,dataset.iloc[:,11:]],axis=1)
treatment.head()
```

```python
import seaborn as sns
```

```python
sns.boxplot(data=[treatment.Treatment1,treatment.Treatment2,treatment.Treatment3,treatment.Treatment4,treatment.Treatment5])
plt.xlabel('treatment i')
plt.ylabel('x pression')
```

```python
gene10treat=treatment.iloc[9,:]
gene10treat.iloc[1:].mean()
```

```python
gene10contr=control.iloc[9,:]
gene10contr.iloc[1:].mean()
```

```python
sns.boxplot(data=[gene10treat[1:],gene10contr[1:]])
plt.xticks([0,1],['treatment','control'])
plt.show()
```

```python
dataset['treatmentmean']=treatment.iloc[:,1:].mean(axis=1)
dataset['controlmean']=control.iloc[:,1:].mean(axis=1)
dataset.head()
```

```python
dataset['modvalue']=abs(dataset['treatmentmean']-dataset['controlmean'])
dataset.head()
```

```python
datamax=dataset.sort_values(by='modvalue',ascending=False)
datamax10=datamax.head(10)
datamax10
```

```python
sns.heatmap(datamax10.iloc[:,1:])
plt.show()
```

```python
gene25treatment=treatment.loc[25]
#gene25treatment
gene25control=control.loc[25]
#gene25control
```

```python
sns.violinplot(data=[gene25treatment[1:],gene25control[1:]])
plt.xticks([0,1],['treatment','control'])
plt.ylabel('value')
plt.show()
```

```python
sns.distplot(treatment.iloc[14,1:],label='treatment')
sns.distplot(control.iloc[14,1:],label='control')
plt.xlabel('distribution')
plt.ylabel('density')
plt.legend()
plt.show()
```

```python
dataset['log2fold']=(dataset['treatmentmean']/dataset['controlmean'])
dataset.head()
```

```python
from scipy import stats
ttest,pvalue=stats.ttest_ind(dataset.iloc[:,1:11],dataset.iloc[:,11:21],axis=1)
dataset['ttest']=ttest
dataset['pvalue']=pvalue
dataset.head()
```

```python

plt.scatter(dataset.log2fold,-dataset.pvalue,c=dataset.pvalue<=0.05)
plt.show()
```
